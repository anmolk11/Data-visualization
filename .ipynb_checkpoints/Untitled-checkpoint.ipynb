```python
import random
x = [i for i in range(1,11)]
y = [random.randint(1,100) for i in range(1,11)]

plt.plot(x,y)
plt.axis([1,10,1,100]) # [xmin,xmax,ymin,ymax]
plt.title("Random plot")
plt.xlabel("Points")
plt.ylabel("Random int")
plt.show()
```
