# Scatter Plot

```python
sns.scatterplot(y = "Salary",x = "EmpSatisfaction",data = df,hue = "Sex")
```

```python
# df["DOB"] = pd.to_datetime(df["DOB"])
# print(type(df.iloc[0]["DOB"]))
print((df.iloc[0]["DOB"]))
```

```python
df['DOB'] = pd.to_datetime(df['DOB'], format= '%m/%d/%y')

# Extract birth year
df['birth_year'] = df['DOB'].dt.year

# Add current date column
df['current_date'] = pd.to_datetime('today')

# Calculate age
df['Age'] = df['current_date'].dt.year - df['birth_year']
```

```python
sns.scatterplot(x = "Salary",y = "Age",data = df,hue = "Sex")
```

```python
sns.scatterplot(x = "Salary",y = "Age",data = df,hue = "Sex",style = "PerformanceScore",alpha = 0.5)
```
