```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

df = pd.read_csv("HRDataset_v14.csv")
print(df.columns)
# style.use("ggplot")
```

# Line Plot

```python
sns.lineplot(x = "EmpSatisfaction",y = "Salary",data = df)
plt.show()
```

```python
sns.lineplot(x = "EmpSatisfaction",y = "Salary",
             data = df,hue = "Sex",palette = "Accent")
plt.show()
```

```python
sns.lineplot(x = "EmpSatisfaction",y = "Salary",data = df)
sns.set(style = "ticks")
plt.figure(figsize = (16,9))
plt.show()
```

# Histogram / Distplot / KDE plot

```python
 sns.distplot(df["Salary"])
```

```python
from scipy.stats import norm
sns.distplot(df["Salary"],fit = norm)
```

```python
 sns.displot(df["Salary"])
```

```python
 sns.histplot(df["Salary"])
```

```python
sns.kdeplot(df["Salary"])
```

# Barplot

```python
print(df.columns)
```

```python
plt.figure(figsize = (20,12))
sns.barplot(x = df["RecruitmentSource"],y = df["Salary"],hue = df["Sex"])
plt.show()
```

```python
plt.figure(figsize = (20,12))
sns.barplot(x = "RecruitmentSource",y = "Salary",hue = "Sex",data = df)
plt.show()
```

```python
#estimator for using diff values for each x axis category (default is mean)
plt.figure(figsize = (20,12))
sns.barplot(x = df["RecruitmentSource"],y = df["Salary"],hue = df["Sex"],estimator = np.median)
plt.show()
```

```python
#ci(confidence interval for the black line)
plt.figure(figsize = (20,12))
sns.barplot(x = "RecruitmentSource",y = "Salary",hue = "Sex",data = df,ci = 75)
plt.show()
```

```python
 plt.figure(figsize = (20,12))
sns.barplot(y = df["RecruitmentSource"],x = df["Salary"],hue = df["Sex"])
plt.show()
```

```python
sns.barplot(x = "Salary",y = "EmpSatisfaction",hue = "Sex",data = df,orient = "h")
```

```python
plt.figure(figsize = (20,12))
sns.barplot(x = "RaceDesc",y = "Salary",hue = "Sex",data = df,palette = "magma")
plt.show()
```

```python
plt.figure(figsize = (20,12))
sns.barplot(x = "RaceDesc",y = "Salary",hue = "Sex",data = df,dodge = False)
plt.show()
```

```python
plt.figure(figsize = (16,9))
kwargs = {'alpha' : 0.8,'linestyle' : ':','linewidth' : 5,'edgecolor' : 'k'}
sns.barplot(x = "RaceDesc",y = "Salary",hue = "Sex",data = df,**kwargs)
```
